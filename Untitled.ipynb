Area Price Prediction

```python
import pandas as pd
```

```python
df=pd.read_csv("areaprice071.csv")
```

```python
df
```

```python
x=df[["area"]]
```

```python
x
```

```python
y=df.price
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(x,y, color="violet",s=100,ec="purple")
plt.show()
```

```python
pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
pred=model.predict(x)
```

```python
df
```

```python
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(x,y, color="violet",s=100,ec="purple")
plt.plot(x,pred,color="green")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
myarea=int(input("Enter the Value of Area : "))
myprice=model.predict([[myarea]])
print("Price is : ",myprice[0])
```
